### __1- How can we add new data in Tableau? (excel, csv, pdf etc.) :__

You can upload files with “Connect” in Tableau. You can upload an excel file, csv file, pdf file etc. or you can connect to a server. When you open Tableau you’ll see a page that is for adding new data and check on your works. You can click the true menu for adding your data.

### __2- To which things we should pay attention while we are adding new data?__ 

Firstly, you should check all columns at source database and Tableau. Are there all columns in the data source menu? If they are, you should check their types. If their types are wrong you should change with true types. 
You can change types from type’s mark option. 

You should check row counts for each table. It is very important while joining tables. Joining tables may cause duplicate rows which is problematic at reporting and analysis.

### __5- What are the dimensions and the measures?__

Tableau separates columns of the data into two groups dimensions and measures. Dimensions contain categorical variables like string, date or geographic values. Measures contain numeric variables. Dimensions are independent  variables. Measures are dependent  variables.  Dimensions’s colour is blue, measures’s colour is green. You can easily use measure value on dimensions or vice versa.

### __11-What are aggregation and granularity?__ 

Whenever you add a measure to your working area, an aggregation is applied to that measure by default. This default property is controlled by the Aggregate Measures setting in the Analysis menu.

If you want to see all of the marks in the view at the most detailed level of granularity, you can disaggregate the view. Disaggregating your data means that Tableau will display a separate mark for every data value in every row of your data source.

### __12- Creating Scatter plot__

Scatter plots’s primary usage is to observe and show relationships between two numeric variables. The dots in a scatter plot not only report the values of individual data points, but also show patterns of the relationship for example positive or negative relationship. So, a scatter plot can also be useful for identifying patterns in data set. We can divide data points into groups based on how closely sets of points cluster together. Scatter plots can also show if there are any unexpected gaps in the data and if there are any outlier points. This can be useful if we want to segment the data into different parts, like in the development of user personas. ### __13- How can we work with multiple worksheets?__

Actually, the real question is how we can connect multiple worksheets with filters? If one of them has a filter how can we add this filter in another worksheets? 

If you have a bar chart with a filter on it and you also want to use the same filter on other charts or maps, you have two options,
 
One way is, you can create the same filter on other charts. If you use this option, you can’t connect charts or maps on the same filter. So each chart or map will not be filtered together. 

Second way is you can use the same filter on all charts and maps. If you use this optıion, you can connect charts and maps on the same filter. For example, if you choose 2014 year value at a chart, other chart’s year will be filtered as the first chart filter. 

If you want to use this feature you should click on filter at one of the charts. And then you should click ‘Apply to worksheet’. Many options will be opened. You can choose whatever you want. 

### __14- How can we join tables at Tableau?__

If we have two or more different data tables or files, we can use join or data blending. Joining is very easy at Tableau. For example, we have two excel pages and we want to join them in Tableau. First step is, we should add the file in a normal way. And then we should drag and drop one of the pages on the data working area. After this, you should drag and drop the other page to the left side of the first page. Which one we choose first table is so important. Because if you want to do left join you must select the page or table to be the main data source which has all the data we want to see. 

When you drag and drop all pages or tables, Tableau will do inner join as default optıion. But you can change the join type when you click on the mark which shows join options. And if you want to change primary keys that are being used for joining or you want to add new keys this section will help you to do that.  

Join types can be summarized as following;

(INNER) JOIN: Returns records that have matching values in both tables
LEFT (OUTER) JOIN: Returns all records from the left table, and the matched records from the right table
RIGHT (OUTER) JOIN: Returns all records from the right table, and the matched records from the left table
FULL (OUTER) JOIN: Returns all records when there is a match in either left or right table

🔺Use joins when;
➻Combining data at the row level. 

### __15- What is Data Blending?__

If you have a table that was joined and if you want to join this table with a new table you should use data blending. Because we may have duplicate values with join action,. 
A blend is like a smart join "on the fly". Data blending is a left join. 
Common fields with the same name are picked up automatically as the blend clause. 
Blending occurs at the granularity of your view. 
Aggregation happens before the blend. 
Blend is unique to each worksheet. 

🔺Use blendings when; 
 ➻Datasource have different levels of granularity
 ➻Datasource come from different systems. 

For making data blending; 
You should join two tables with normal joining. And then you should click Tableau’s mark under field for adding a new table. And you should select your new file. After that, you will see two different data sources at your worksheet. 

You should select the data source that you want to use. After you select, you will see columns of this data source. If you want to see relationships and add new relationships you should follow the steps below. 

➻ Data ⟶ Edit relationships ⟶ A page will open ⟶ You can see all relationships ⟶ if you select custom, the add button will be active ⟶ new page will open ⟶ you can add new relationships ⤵
